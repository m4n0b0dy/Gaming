```python
meta_data = []
with open('C:/Users/kvenuti/Music/data/meta_data.json') as f:
    for line in f:
        meta_data.append(ast.literal_eval(line))
```

```python
c = 0
titles = []
for t in meta_data:
    if 'title' in t.keys():
        c+=1
        titles.append(t['title'])
print(c,c/len(meta_data))
```
